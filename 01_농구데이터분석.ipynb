### 데이터 전처리 체이닝처리

```python
a1 = {
    'Date' : '경기일자',
    'Start (ET)' : '시작시간',
    'Visitor/Neutral' : '방문팀',
    'PTS' : '방문팀점수',
    'Home/Neutral' : '홈팀',
    'PTS.1' : '홈팀점수',
    'Unnamed: 6' : 'Box',
    'Unnamed: 7' : '연장전여부',
    'Attend' : '관중수',
    'Notes' : 'Notes'
}

# 날짜를 원하는 양식으로 만들어 다시설정
# 경기날짜를 인덱스로 추가
# 원하는 컬럼만 가져오기
games_all = pd.read_csv('data/nav_2018.csv') \
                .rename(columns=a1,)
games_all = games_all\
[['경기일자', '방문팀', '방문팀점수', '홈팀', '홈팀점수']]\
.assign(경기일자=lambda x: pd.to_datetime(x['경기일자'], format\
                                      = '%a, %b %d, %Y'))\
.set_index('경기일자',append = True)\
.rename_axis(['게임','경기일자'])
```

```python
games_all
```

```python
# 인덱스를 다시설정한다
a1=games_all.reset_index()
a1
```

```python
tidy = pd.melt(a1,id_vars=['게임','경기일자'],value_vars = ['방문팀','홈팀'])
```

```python
tidy.query('경기일자 == "2017-10-17"')
```

```python
tidy.head()
```

```python
# 날짜별 경기수
# 경기일자 타입을 파악
tidy.경기일자.dtype
```

```python
tidy.groupby('value')['경기일자'].diff()
```

```python
tidy.rename(columns = {'value':'팀','variable':'팀구분'},inplace = True)
```

```python
# 팀별 경기일자
tidy.groupby('팀')['경기일자'].count()
```

```python
# nan : not a number
# nat : not a Time
# 첫번쨰로우는 무조건 결측치
tidy.groupby('팀')['경기일자'].diff()
```

```python
a1 = tidy.groupby('팀')['경기일자'].diff().dt.days==1
tidy['경기일자'][a1]
```

```python
ab = tidy.groupby('팀')['경기일자'].diff().dt.days -1
```

```python
ab.shape
```

```python
ab.isnull().sum()
```

```python
# 경기일자를 기준으로 정렬
a1 = tidy.sort_values('경기일자')
# 팀을 기준으로 그룹을 묶는다
a2 = a1.groupby('팀')
# 경기일자를 기준으로 차이값을 구한다
a3 = a2.경기일자.diff()
# a3에서 -1해서 휴식일을 구한다
tidy['휴식일'] = a3.dt.days-1
```

```python
a2['경기일자']
```

```python
tidy.isna().sum()
```

```python
# 결측치에 0을 셋팅한다
tidy = tidy.fillna(0)
tidy
```

```python
# pivot테이블을 구한다
by_game = pd.pivot_table(tidy, values = '휴식일',
                        index = ['게임','경기일자'],
                        columns = '팀구분')
```

```python
by_game
```

```python
# 컬럼 이름 변경
a1 = {
    '방문팀':'방문팀 휴식일',
    '홈팀':'홈팀휴식일'
}

by_game = by_game.rename(columns = a1)
by_game
```

```python
# 원본과 합친다
df = pd.concat([games_all, by_game], axis = 1)
df
```
